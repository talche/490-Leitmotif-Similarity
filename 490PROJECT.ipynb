```python
import tensorflow as tf
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import librosa
from tensorflow.keras.backend import clear_session
import os

# Dictionary specifying the segment duration (in seconds) for each song audio 1 which is Undertale OST: 015 - Sans
# Has a segment length of two which we have chosen by listening by ear
SEGMENT_DURATIONS = {
    'audio1.mp3': 3,
    'audio2.mp3': 3,
    'audio3.mp3': 2,
}

def load_audio(file_path, default_segment_duration=2):
    file_name = os.path.basename(file_path)
    segment_duration = SEGMENT_DURATIONS.get(file_name, default_segment_duration)
    y, sr = librosa.load(file_path, sr=None)
    segment_length = int(segment_duration * sr)
    print(f"File: {file_name} | Sample Rate: {sr} Hz | Segment Duration: {segment_duration} s | Segment Length: {segment_length} samples")
    segments = [y[i:i + segment_length] for i in range(0, len(y), segment_length)
                if len(y[i:i + segment_length]) == segment_length]
    scaler = StandardScaler()
    segments = [scaler.fit_transform(segment.reshape(-1, 1)).flatten() for segment in segments]
    return np.array(segments, dtype='float16'), sr

def build_autoencoder(input_dim, encoding_dim=32):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(128, activation=tf.keras.activations.relu, kernel_regularizer=l2(0.01))(input_layer)
    encoded = Dropout(0.3)(encoded)
    encoded = Dense(64, activation=tf.keras.activations.relu, kernel_regularizer=l2(0.01))(encoded)
    encoded = Dropout(0.3)(encoded)
    encoded_output = Dense(encoding_dim, activation=tf.keras.activations.relu)(encoded)

    decoded = Dense(64, activation=tf.keras.activations.relu, kernel_regularizer=l2(0.01))(encoded_output)
    decoded = Dense(128, activation=tf.keras.activations.relu, kernel_regularizer=l2(0.01))(decoded)
    decoded_output = Dense(input_dim, activation='sigmoid')(decoded)

    autoencoder = Model(inputs=input_layer, outputs=decoded_output)
    encoder = Model(inputs=input_layer, outputs=encoded_output)

    autoencoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='mse')
    return autoencoder, encoder

def train_autoencoder(autoencoder, segments, batch_size=32, epochs=100, validation_split=0.2):
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = autoencoder.fit(segments, segments,
                              epochs=epochs,
                              batch_size=batch_size,
                              shuffle=True,
                              validation_split=validation_split,
                              verbose=1,
                              callbacks=[early_stopping])
    return history

def plot_loss_curve(history):
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title('Loss Curve')
    plt.show()

def visualize_embeddings(embeddings):
    tsne = TSNE(n_components=2, random_state=42, perplexity=1)
    tsne_embeddings = tsne.fit_transform(embeddings)
    plt.scatter(tsne_embeddings[:, 0], tsne_embeddings[:, 1], s=5)
    plt.title('t-SNE Visualization of Latent Embeddings')
    plt.show()

def calculate_similarity_matrix(embeddings):
    similarity_matrix = 1 - pairwise_distances(embeddings, metric="cosine")
    plt.imshow(similarity_matrix, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Cosine Similarity Matrix')
    plt.show()
    return similarity_matrix

def detect_leitmotif(embeddings, leitmotif_embedding, threshold=0.8):
    similarities = [1 - pairwise_distances([e], [leitmotif_embedding], metric="cosine")[0, 0] for e in embeddings]
    leitmotif_regions = []
    in_region = False
    start_index = None

    for i, sim in enumerate(similarities):
        if sim >= threshold:
            if not in_region:
                in_region = True
                start_index = i
        else:
            if in_region:
                in_region = False
                leitmotif_regions.append((start_index, i - 1))

    if in_region:
        leitmotif_regions.append((start_index, len(similarities) - 1))

    return leitmotif_regions, similarities

def plot_leitmotif_detection(similarities, leitmotif_regions, threshold):
    plt.plot(similarities)
    plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
    for start, end in leitmotif_regions:
        plt.axvspan(start, end, color='green', alpha=0.3)
    plt.title('Leitmotif Similarity Over Time')
    plt.xlabel('Segment Index')
    plt.ylabel('Similarity')
    plt.legend()
    plt.show()

def main(file_paths):
    leitmotif_embeddings = []
    for file_path in file_paths:
        print(f"Processing file: {file_path}")
        segments, sr = load_audio(file_path)

        autoencoder, encoder = build_autoencoder(input_dim=segments.shape[1])
        history = train_autoencoder(autoencoder, segments)
        plot_loss_curve(history)

        embeddings = encoder.predict(segments)
        visualize_embeddings(embeddings)

        similarity_matrix = calculate_similarity_matrix(embeddings)

        leitmotif_segment = segments[0]
        leitmotif_embedding = encoder.predict(np.array([leitmotif_segment], dtype='float16'))

        leitmotif_regions, similarities = detect_leitmotif(embeddings, leitmotif_embedding[0])
        plot_leitmotif_detection(similarities, leitmotif_regions, threshold=0.8)

        segment_duration = SEGMENT_DURATIONS.get(os.path.basename(file_path), 2)
        leitmotif_times = [(start * segment_duration, (end + 1) * segment_duration) for start, end in leitmotif_regions]
        print("Detected leitmotif times (in seconds):", leitmotif_times)

        leitmotif_embeddings.append((os.path.basename(file_path), leitmotif_embedding[0]))

    for i in range(len(leitmotif_embeddings)):
        for j in range(i + 1, len(leitmotif_embeddings)):
            song1, embedding1 = leitmotif_embeddings[i]
            song2, embedding2 = leitmotif_embeddings[j]
            similarity = 1 - pairwise_distances([embedding1], [embedding2], metric="cosine")[0, 0]
            print(f"Similarity between {song1} and {song2}: {similarity}")

if __name__ == "__main__":
    audio_dir = "./"
    file_paths = [os.path.join(audio_dir, file) for file in os.listdir(audio_dir) if file.endswith(('.mp3', '.wav'))]
    if not file_paths:
        print("No audio files found in the current directory.")
    else:
        main(file_paths)
```
